## Matplotlib example workflow

```python
# import the matplotlib and get ready for Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# Make the data ready to use
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
#Setup the plot ...
fig, ax = plt.subplots()
# 3. Plot data
ax.plot(x,y)
# 4.Costomize the plot we have created
ax.set(title= "Simple plot", xlabel="X-axsis", ylabel="Y-lable")
fig.savefig("images/theFirstPlot.jpg")
```

```python
# Making a plot from a dictinary
nut_butter_pices = {"Almost bettert": 11,
                    "Peaunt butter": 8,
                    "Cashew butter": 12,}

fig, ax = plt.subplots()
ax.bar(nut_butter_pices.keys(), height= nut_butter_pices.values())
ax.set(title= " My nuts Store", ylabel="Price in $")
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_pices.keys()), list(nut_butter_pices.values()))
```

```python
myNumber = np.random.randn(1000)
fig, ax = plt.subplots()
fig.hist(myNumber)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(20, 10))
ax1.plot(x, x/2)
ax2.scatter(x,np.sin(x))
ax3.bar(nut_butter_pices.keys(), height=nut_butter_pices.values())
ax4.hist(np.random.randn(1000))
```

```python
fig, ax = plt.subplots(ncols=2,nrows=2,figsize=(20,10))
ax[0,0].scatter(np.random.random(10), np.random.random(10))
ax[0,1].plot(x , x**2)
ax[1,0].hist((np.random.rand(200)))
ax[1,1].barh(list(nut_butter_pices.keys()), list(nut_butter_pices.values()))
ax.set(title="The secound wat ro demonstrate")
```
